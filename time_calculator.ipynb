```python
def add_time(start, duration, day=''):

    meridiem_flip = {'AM': 'PM', 'PM':'AM'}
    days_of_week = {"monday": 1, "tuesday": 2, "wednesday": 3, "thursday": 4, "friday": 5, "saturday": 6, "sunday": 7}
    list_days_of_week = ['', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday' ]
    
    # Splitting the parameters
    [timeStart, meridiem] = start.split(' ')
    [hourStart, minuteStart] = timeStart.split(':')
    [hourDuration, minuteDuration] = duration.split(':')
    
    # Converting to integers    
    hourStart = int(hourStart)
    minuteStart = int(minuteStart) 
    hourDuration = int(hourDuration) 
    minuteDuration = int(minuteDuration)
    
    # Calculating the minutes
    new_minutes = minuteStart + minuteDuration
    if new_minutes >= 60:
        hourDuration += 1
        new_minutes %= 60
    if 0 <= new_minutes < 10:
        new_minutes = '0' + str(new_minutes)
     
    # Calculating the hours and the days
    day_counts = 0
    hourCount = hourStart + hourDuration
        
    # No changes of the meridiem      
    if hourCount < 12:
        new_hour = hourCount
        
    # With change of meridiem and 'next day'for PM
    if hourCount >= 12 and hourCount < 24:
        new_hour = hourCount - 12
        if meridiem == 'PM':
            day_counts = 1
        meridiem = meridiem_flip[meridiem]
        
    # With change of days to 'AM' and 'PM''
    if hourCount >= 24:
        [day_counts, new_hour] = divmod(hourCount, 24) 
        if new_hour >= 12 and meridiem == 'PM':
            day_counts += 1
        if new_hour >= 12:
            new_hour = new_hour - 12
            meridiem = meridiem_flip[meridiem]   
    
    # New_hour = 0
    if new_hour == 0:
        new_hour = 12
    
    # Returning the new_time
    new_time = f'{new_hour}:{new_minutes} {meridiem}'
    
    
    # Counting the days 
    #  No change of day  
    if day_counts == 0 and day != '':
        new_time += f', {day.capitalize()}'
        
    # Only next day
    elif day_counts == 1 and day == '':
            new_time += ' (next day)'
    
    # Only count days       
    elif day_counts > 1 and day == '':
            new_time = f'{new_hour}:{new_minutes} {meridiem} ({day_counts} days later)'
    
    # One count day and the new day of week
    elif day_counts == 1 and day !='':  
        passed_days = days_of_week[day.lower()] + day_counts
        # Less than one week
        if passed_days <= 7:
            day = list_days_of_week[passed_days]
        # More than one week
        elif passed_days > 7:
            day = list_days_of_week[passed_days]
        new_time += f', {day} (next day)' 
        
    
    # More than one count day and the new day of week
    elif day_counts > 1 and day !='':  
        passed_days = days_of_week[day.lower()] + day_counts       
        # Less than one week
        if passed_days <= 7:
            day = list_days_of_week[passed_days]
        # A week 
        elif passed_days == 7:
            day = day.capitalize()
        # More than one week
        elif passed_days > 7:
            passed_days %= 7
            day = list_days_of_week[passed_days]
        new_time += f', {day} ({day_counts} days later)' 
    
    # Result     
    return new_time

print(add_time("2:59 AM", "24:00", "saturDay"))
```
